```python
import pandas as pd
import os
from scipy import stats
from scipy.stats import spearmanr
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import statsmodels.formula.api as smf
```

```python
file_path = "/Users/remyfernando/Desktop/Noguchi Active/BSc Psychology/Year 3/Research Project/Project data/Qualtrics_ready_for_analysis.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
# Create a new column for z-scores and assign values based on device type
df['z_pickups'] = None
df.loc[df['ios_pickups'].notna(), 'z_pickups'] = stats.zscore(df['ios_pickups'].dropna())
df.loc[df['non_ios_avg_pickups'].notna(), 'z_pickups'] = stats.zscore(df['non_ios_avg_pickups'].dropna())
df["z_pickups"] = pd.to_numeric(df["z_pickups"], errors='coerce')
```
