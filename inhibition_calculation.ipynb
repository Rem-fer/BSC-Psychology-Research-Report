# Inhibition score calculation
In this notebook, raw data from the PsychoPy Go/No-Go task is processed and the inhibition score calculated for one participant.
For this project, inhibition scores were calculated by counting correct responses on No-Go trials (i.e. withholding a response when the letter X appeared).


```python
import pandas as pd
import os
```

```python
# Define the file path to the CSV 
file_path = '/Users/remyfernando/Desktop/Noguchi Active/BSc Psychology/Year 3/Research Project/Project data/Psychopy raw CSVs/AG115_New_Go_No_go_task_2024-02-09_12h00.50.188.csv'

df = pd.read_csv(file_path)
df.head()
```

```python
# Get participant ID from the file name
participant_id = os.path.basename(file_path).split('_')[0]
```

```python
# Get the index of the first non-NA value in the "ready.started" column
## This is used as starting point for the experiment, as the rows before ar ejust a testing phase
starting_index = df.index[df["ready.started"].notna()].to_list()[0]
```

```python
# Select only the relevant columns for analysis
relevant_coluns = ['corrAns', 'letter', 'response.keys', 'response.corr']
df_relevant = df[relevant_coluns].iloc[starting_index + 1:]

# Rename the columns to more descriptive names
df_relevant.rename(columns={
    'corrAns': 'correct_answer',
    'letter': 'letter',
    'response.keys': 'response_keys',
    'response.corr': 'response_correct'
}, inplace=True)

pd.set_option('display.max_rows', None)
df_relevant.head(20)
```

```python
# Filter for No-Go trials (letter == 'X') where the participant correctly withheld their response (response.corr == 1)
correct_responses = df_relevant[(df_relevant['letter'] == 'X') & (df_relevant['response_correct'] > 0)]
score = len(correct_responses)
print(f"{participant_id} scored {score} out of 100")
```
