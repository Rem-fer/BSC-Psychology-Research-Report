### Normality test

```python
# Normality tests
columns_to_analyse = ['maas_score', 'ios_pickups', 'non_ios_avg_pickups', 'inhibition_score']

normality_results = {}
for col in columns_to_analyse:
    stat, p_value = stats.shapiro(df[col].dropna())
    normality_results[col] = {
        "skewness": round(df[col].skew(), 2),
        "kurtosis": round(df[col].kurtosis(), 2),
        "shapiro_statistic": round(stat, 4),
        "shapiro_p_value": round(p_value, 4),
        "normal": p_value > 0.05
    }
normality_df = pd.DataFrame(normality_results).T
print(normality_df)
```

### Normality & Outlier Assessment Conclusion

Descriptive statistics and Shapiro-Wilk tests were conducted for each variable prior to analysis.

MAAS scores were approximately normally distributed (M = 3.91, SD = 0.69, skewness = 0.40, kurtosis = 1.09), confirmed by a non-significant Shapiro-Wilk test (W = 0.957, p = .249).

iOS pickup frequency was also normally distributed (M = 124.64, SD = 65.82, skewness = -0.36, kurtosis = -0.73), with a non-significant Shapiro-Wilk test (W = 0.950, p = .557).

Inhibition scores were normally distributed (skewness = 0.06, kurtosis = -0.60), confirmed by a non-significant Shapiro-Wilk test (W = 0.975, p = .653).

Non-iOS average pickups violated the assumption of normality (M = 59.65, SD = 34.97, skewness = 1.95, kurtosis = 4.33), with a significant Shapiro-Wilk test (W = 0.774, p = .001). Visual inspection of the distribution confirmed positive skew. One outlier was identified but retained as the Z-score fell within the acceptable ±3 range (Osborne & Overbay, 2004), and removal did not resolve the normality violation. Given the correlational design, slight deviations from normality were considered tolerable and analyses proceeded accordingly.

### Z-Score Standardisation

Phone pickup scores were standardised using Z-scores to account for systematic differences in pickup frequency between device types (iPhone vs Android/Samsung). iPhone users' pickup data comes from iOS screen time tracking, while Android and Samsung users self-reported daily pickups — these different measurement methods introduce scale differences. By calculating Z-scores separately within each device group, scores are placed on a consistent scale, allowing valid comparisons across all participants.

```python
# Create a new column for z-scores and assign values based on device type
df['z_pickups'] = None
df.loc[df['ios_pickups'].notna(), 'z_pickups'] = stats.zscore(df['ios_pickups'].dropna())
df.loc[df['non_ios_avg_pickups'].notna(), 'z_pickups'] = stats.zscore(df['non_ios_avg_pickups'].dropna())
```

## 4. Inferential analysis

```python
# Convert z_pickups to numeric
print(df[['z_pickups', 'maas_score', 'inhibition_score']].dtypes)

df["z_pickups"] = pd.to_numeric(df["z_pickups"], errors='coerce')
```

### Multile regression
Maas score and Inhibition as predictors with phone pickus as outcome

```python
# Multiple regression, 
model1 = smf.ols('z_pickups ~ maas_score + inhibition_score', data=df).fit()
print(model1.summary())
```

### Multiple Regression Results

A multiple regression was conducted with MAAS score and inhibition score as predictors of standardised phone pickup frequency (z_pickups).

The overall model was not statistically significant (R² = 0.124, Adj. R² = 0.061, F(2, 28) = 1.98, p = .157), suggesting that MAAS and inhibition scores together explained only 12.4% of the variance in phone checking behaviour.

MAAS score was not a significant predictor (β = 0.04, t = 0.15, p = .884). Inhibition score approached but did not reach significance (β = -0.03, t = -1.86, p = .073), with the negative coefficient suggesting that higher inhibitory control was associated with fewer phone pickups — a direction consistent with theoretical expectations.

The Durbin-Watson statistic (1.892) indicated no substantial autocorrelation in the residuals. The Omnibus test was significant (p = .043), indicating mild non-normality in the residuals, consistent with the normality violations identified during descriptive analysis.

### Simple regression

```python
# Simple regression
model2 = smf.ols('inhibition_score ~ maas_score', data=df).fit()
print(model2.summary())
```

```python
fig3 = px.scatter(df, x='maas_score', y='inhibition_score', trendline='ols', title='MAAS vs Inhibition')
fig3.show()
```

### Simple Regression Results

A simple linear regression was conducted with MAAS score as a predictor of inhibition score.

The model was statistically significant (R² = 0.178, Adj. R² = 0.150, F(1, 29) = 6.27, p = .018), with MAAS score explaining 17.8% of the variance in inhibition score.

MAAS score was a significant positive predictor of inhibition (β = 8.82, t = 2.51, p = .018), suggesting that higher trait mindfulness was associated with better inhibitory control.

The Durbin-Watson statistic (1.182) indicates some positive autocorrelation in the residuals, which should be interpreted with caution. However, the Omnibus test was non-significant (p = .976), suggesting residuals were approximately normally distributed.

### Correlation Table

```python
relevant_columns = ['maas_score', 'inhibition_score', 'z_pickups']
correlation_matrix = df[relevant_columns].corr(method='spearman')
print(correlation_matrix)
```

```python

correlations = {}
for col in relevant_columns:
    for col2 in relevant_columns:
        if col != col2:
            corr, p = spearmanr(df[col], df[col2])
            correlations[(col, col2)] = (corr, p)
correlation_matrix = pd.DataFrame(correlations, index=['spearman_correlation', 'p_value']).T
print(correlation_matrix)
```

## 5. Conclusion

**H1 — Not supported.** MAAS score was not a significant predictor of phone checking frequency in the multiple regression (β = 0.04, p = .884).

**H2 — Not supported.** Inhibition score did not significantly predict phone checking frequency, though it approached significance (β = -0.03, p = .073), with the direction of effect consistent with the hypothesis.

**H3 — Supported.** MAAS score significantly predicted inhibitory control (β = 8.82, p = .018), suggesting that higher trait mindfulness is associated with better inhibitory control.
