## Data cleaning

```python
file_name = "Qualtrics_raw.csv"
df = pd.read_csv(f"/Users/remyfernando/Desktop/Noguchi Active/BSc Psychology/Year 3/Research Project/Project data/{file_name}")
```

### Changing device type values from numerical to actual Names

```python
# First converting column to numeric
df["device"] = pd.to_numeric(df["device"], errors='coerce')
df["device"].dtypes

# Then changing device types to names
df['device'] = df['device'].astype(object)
df.loc[df['device'] == 1, 'device'] = 'iPhone'
df.loc[df['device'] == 2, 'device'] = 'Android'
df.loc[df['device'] == 3, 'device'] = 'Samsung'

df.head(20)
```

## Cleaning pick-up columns
There were two different set of weekdays day, one for Android and. one for Samsung. This was because of Qualtrics set-up which collected differently based on device.
Merging both Samsung and Android data into the same 7 columns for each days collected

```python
print("Samsung")
print(df[df['device'] == 'Samsung'][['screen_time_mon', 'pickups_mon']].head(3))
print("android")
print(df[df['device'] == 'Android'][['screen_time_mon', 'pickups_mon']].head(3))
```

```python
days = ['mon', 'tue', 'wed', 'thu', 'fri', 'sat', 'sun']

for day in days:
    df.loc[df['device'] == 'Samsung', f'pickups_{day}'] = df.loc[df['device'] == 'Samsung', f'screen_time_{day}']
```

```python
screentime_cols = [col for col in df.columns if 'screen_time' in col]
print(screentime_cols)
```

```python
df.drop(columns=screentime_cols, inplace=True)
df.columns
```

```python
questionaire_columns = [ 'Questionnaire_1',
       'Questionnaire_2', 'Questionnaire_3', 'Questionnaire_4',
       'Questionnaire_5', 'Questionnaire_6', 'Questionnaire_7',
       'Questionnaire_8', 'Questionnaire_9', 'Questionnaire_10',
       'Questionnaire_11', 'Questionnaire_12', 'Questionnaire_13',
       'Questionnaire_14', 'Questionnaire_15',
       ]
print(questionaire_columns)
```

```python
df.head()
```

### Calculating and adding Non-IOS device pick- average column

```python
days = ['pickups_mon', 'pickups_tue', 'pickups_wed', 'pickups_thu', 'pickups_fri', 'pickups_sat', 'pickups_sun']
 # Convert the pickups columns to numeric, coercing errors to NaN
for day in days:
    df[day] = pd.to_numeric(df[day], errors='coerce')

df["non_ios_avg_pickups"] = df[days].mean(axis=1).round()

# Rounding MAAS scores to 2 decimal places
df["maas_score"] = pd.to_numeric(df["maas_score"], errors='coerce').round(2)

df.head(20)
```

```python
df.columns
```

```python
# Getting rid of columns that are not relevant to the analysis
relevant_columns = ['participant_id',
       'gender', 'occupation', 'education', 'age', 'phone_checking_frequency',
       'device', 'ios_pickups',
       'maas_score', 'non_ios_avg_pickups']

df = df[relevant_columns]
df.head(20)
```

```python
# Renaming the phone_checking_frequency column to more_than_one_device 
df.rename(columns={'phone_checking_frequency': 'more_than_one_device'}, inplace=True)
df.head()
```

### Df Backup

```python
# Creating a backup of the DataFrame before making further changes
df_backup = df.copy() 
```

```python
# Checking the unique values in the more_than_one_device column
print(df['more_than_one_device'].unique())
```

```python
# Converting the more_than_one_device column to boolean values
df['more_than_one_device'] = df['more_than_one_device'].astype(object)
df.loc[df['more_than_one_device'] == "1", 'more_than_one_device'] = False
df.loc[df['more_than_one_device'] == "2", 'more_than_one_device'] = True
df.head()
```

```python
# Converting the gender column to categorical values
df['gender'] = df['gender'].astype(object)
df.loc[df['gender'] == "1", 'gender'] = "Male"
df.loc[df['gender'] == "2", 'gender'] = "Female"
df.head()
```

```python
# Dropping the occupation and education
df.drop(columns=['occupation', 'education'], inplace=True)
```

```python
# Creating a cleaned version of the DataFrame for analysis
qualtrics_cleaned = df.copy()
file_name = "Qualtrics_cleaned.csv"
qualtrics_cleaned.to_csv(os.path.join(OUTPUT_FOLDER_PATH, file_name), index=False)
```

### Fecthing inhibition scores file

```python
# Loading inhibition scores
inhibition_scores_path = "/Users/remyfernando/Desktop/Noguchi Active/BSc Psychology/Year 3/Research Project/Project data/inhibition_scores.csv"
inhibition_scores_df = pd.read_csv(inhibition_scores_path)
inhibition_scores_df.head()
```

### Joining the two table on participant_id

```python
# Standardize participant IDs to lowercase and strip whitespace
inhibition_scores_df["participant_id"] = inhibition_scores_df["participant_id"].str.lower().str.strip()
qualtrics_cleaned["participant_id"] = qualtrics_cleaned["participant_id"].str.lower().str.strip()
```

```python
df_merged = inhibition_scores_df.merge(qualtrics_cleaned, on='participant_id', how='inner')
df_merged.head()
```
